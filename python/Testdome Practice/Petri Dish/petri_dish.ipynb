```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(1)
df = pd.DataFrame({
    "Day Name": [f"Day {i}" for i in range(21)],
    "Day": [i for i in range(21)],
    "coverage A": np.linspace(50, 70, 21) * np.random.uniform(0.9, 1.7, size=21),
    "coverage B": np.linspace(50, 40, 21)
})
```

```python
#Pearson Corellation
df[["Day", "coverage A"]].corr()
```

```python
#Pearson Corellation
df[["Day", "coverage B"]].corr()
```
